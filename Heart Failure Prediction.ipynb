```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/heart.csv')
```

```python
df
```

```python
# Check for null values
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.corr(),cmap='viridis',annot=True)
```

```python
sns.pairplot(data=df,hue='HeartDisease',corner=True,plot_kws={'alpha':0.5})
```

```python
plt.figure(figsize=(8,4),dpi=150)
plt.title('Heart Disease Distribution')
sns.countplot(data=df,x='HeartDisease',hue='Sex',palette='Set1');
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='Cholesterol',kde=True,bins=50,color='red',edgecolor='black')
```

```python
plt.figure(figsize=(8,4),dpi=150)
plt.title('Chest Pain Distribution')
sns.countplot(data=df,x='ChestPainType',hue='Sex',palette='Set2');
```

```python
plt.figure(figsize=(8,4),dpi=150)
plt.title('Resting ECG Distribution')
sns.countplot(data=df,x='RestingECG',hue='Sex',palette='viridis');
```

```python
plt.figure(figsize=(15,10))
for i,col in enumerate(df.columns,1):
    plt.subplot(4,3,i)
    plt.title(f"Distribution of {col} Data")
    sns.histplot(df[col],kde=True)
    plt.tight_layout()
    plt.plot()
```
